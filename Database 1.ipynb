#### Import module, initialize variable, and define function

```python
import pymysql
import pandas as pd

db = pymysql.connect(charset='utf8', unix_socket="/var/lib/mysql/mysql.sock")
cursor=db.cursor(pymysql.cursors.DictCursor)
rst = None

def q(query):
    global rst
    cursor.execute(query)
    rst = cursor.fetchall()
    return(pd.DataFrame(rst))
```
