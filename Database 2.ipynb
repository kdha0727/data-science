MySQL과의 연결을 위한 pymysql 모듈 및 데이터 조회를 위한 pandas 모듈 로드 \
로컬 데이터베이스 접속, SQL문 실행을 위한 함수 정의

```python
# Import module
import pymysql
import pandas as pd

# Initialize variable
db = pymysql.connect(charset='utf8', unix_socket="/var/lib/mysql/mysql.sock")
cursor=db.cursor(pymysql.cursors.DictCursor)
rst = None

# Define function
def q(query):
    global rst
    cursor.execute(query)
    rst = cursor.fetchall()
    return pd.DataFrame(rst)


cursor.execute("use mimic")
```

# DML - 데이터 조작어


###### SQL 기능에 따른 분류
DDL - 데이터 정의어: 테이블이나 테이블간의 관계 정의 (CREATE, ALTER, DROP 문 등) \
DML - 데이터 조작어: 데이터를 검색, 삽입, 수정, 삭제 (SELECT, INSERT, DELETE, UPDATE 문 등) \
DCL - 데이터 제어어: 데이터의 사용 권한을 관리

###### SELECT 문의 구성 요소
select \[all | distinct\] \[속성이름(들) | *\] \
from 테이블이름(들) \
\[where 검색조건(들)\] \
\[group by 속성이름(들)\] \
\[having 검색조건(들)\] \
\[orderby 속성이름(들) [asc | desc] \]

# 부속질의

## 부속질의란?
SQL문 내 또 다른 SQL문 사용

예제 1: 체온(50825)이 가장 높게 측정된 환자 ID 추출

```python
q("""SELECT subject_id FROM labevents WHERE itemid='50825' AND value = (
SELECT max(value) FROM labevents WHERE itemid='50825');""")
```

```python
# 위의 결과와 함께 참고
q("select * from labevents where itemid='50825' order by value desc")
```

예제 2: 체온이 38도 이상인 환자들이 갖고있는 진단명 추출

```python
q("""
SELECT DISTINCT long_title FROM d_icd_diagnoses
    WHERE icd9_code in (SELECT icd9_code FROM diagnoses_icd
        WHERE subject_id in (SELECT subject_id FROM labevents 
            WHERE itemid='50825' and value>=38.0))
""")
```

```python
# 위의 결과와 함께 참고 - 1/2
q("""
SELECT icd9_code FROM diagnoses_icd
    WHERE subject_id in (SELECT subject_id FROM labevents 
        WHERE itemid='50825' and value>=38.0)
""")
```

```python
# 위의 결과와 함께 참고 - 2/2
q("""
SELECT subject_id FROM labevents 
    WHERE itemid='50825' and value>=38.0
""")
```

## 상관 부속질의 (Correlated Subquery)
상위 부속질의의 투플을 이용하여 하위 부속질의를 계산 \
즉, 상위 부속질의와 하위 부속질의가 독립적이지 않고 서로 관련되어 있다

예제: 각 입원별로 가장 최초로 측정한 체온 검사 결과값을 추출

```python
q("""
SELECT * FROM labevents AS L1 WHERE itemid='50825' AND charttime= 
    (SELECT MIN(charttime) FROM labevents AS L2 WHERE itemid='50825' AND L1.hadm_id=L2.hadm_id)
    ORDER BY subject_id, hadm_id;
""")
```

 ### RANK / DENS_RANK / ROW_NUMBER
 OVER 과 함께 사용

```python
# 예제: 각 환자별로 가장 높게 측정한 체온 결과값을 추출
q("""
select 
    *, 
    RANK() OVER (partition by HADM_ID order by value DESC) AS rank,
    DENSE_RANK() OVER (partition BY HADM_ID order by value DESC) AS dense_rank,
    ROW_NUMBER() OVER (partition BY HADM_ID order by value DESC) AS row_num
FROM labevents WHERE itemid='50825' LIMIT 20;
""")
```

## 부속질의의 종류

###### 스칼라 부속질의 - Scalar Subquery: 
SELECT 절에서 사용, \
단일 값을 반환하기 때문에 스칼라 부속질의라고 한다.

###### 인라인 뷰 - Inline View, Table Subquery: 
FROM 절에서 사용, \
결과를 뷰(view) 형태로 반환하기 때문에 인라인 뷰라고 한다.

###### 중첩질의 - Nested Subquery, Predicate Subquery: 
WHERE 절에서 사용, \
WHERE 절에서 술어와 함께 사용되며, 결과를 한정시키기 위해 사용된다. 상관 혹은 비상관 형태다.

### 스칼라 부속질의 (Scalar Subquery)
부속질의의 결과 값을 단일 행, 단일 열의 스칼라 값으로 반환 \
원칙적으로 스칼라 값이 들어갈 수 있는 모든 곳에 사용 가능 \
일반적으로 SELECT 문과 UPDATE SET 절에 사용 \
주질의와 부속질의의 관계는 상관 및 비상관 모두 가능

예제: 진단내역을 진단명과 함께 추출

```python
q("""
select *,
(select short_title from d_icd_diagnoses as icd where dx.ICD9_CODE=icd.ICD9_CODE) as short_title
from diagnoses_icd as dx;
""")  # 상관 부속질의 사용
```

```python
# 부속질의 대신 Join을 사용한다면 아래와 같이 사용 가능. 결과는 같다.
q("SELECT a.*, b.short_title FROM diagnoses_icd AS a, d_icd_diagnoses AS b WHERE a.icd9_code=b.icd9_code")
```

### 인라인 뷰 (Inline View, Table Subquery)
FROM 절에서 사용되는 부속질의 \
보통의 테이블과 같은 형태로 사용 가능 \
부속질의 결과는 다중 행, 다중 열이어도 관계 없음
###### FROM 절에는 테이블 이름이 들어가는 것이 원칙이므로 인라인 뷰에서는 반드시 별칭을 사용해야 한다. 

예제: 환자별로 체온이 38도 이상인 결과값을 가장 많이 갖고 있는 순서대로 환자 추출

```python
q("""
SELECT hadm_id, COUNT(*) AS cnt FROM
        (SELECT * FROM labevents WHERE itemid='50825'
        AND value>=38.0) AS v
GROUP BY hadm_id ORDER BY cnt DESC
;""")  # 인라인 뷰에서는 별칭을 반드시 사용해야 한다
```

### 인라인 뷰 (Inline View, Table Subquery)
WHERE 절에서 사용되는 부속질의 \
데이터를 선택하는 조건과 같이 사용된다.
###### 비교 (=, <, ...)
반환 행 - 단일, 반환 열 - 단일, 상관 - 가능
###### 집합 (IN, NOT IN)
반환 행 - 다중, 반환 열 - 단일, 상관 - 가능
###### 한정  - quantified (ALL, SOME(ANY))
반환 행 - 다중, 반환 열 - 단일, 상관 - 가능
###### 존재 (EXISTS, NOT EXISTS) 
반환 행 - 다중, 반환 열 - 다중, 상관 - 필수

###### 중첩 질의 - IN

```python
q("""
SELECT DISTINCT LONG_TITLE FROM d_icd_diagnoses
WHERE ICD9_CODE IN (SELECT ICD9_CODE FROM diagnoses_icd
        WHERE subject_id IN (SELECT subject_id FROM labevents
                WHERE itemid='50825' AND value>=38.0)
        );
""")
```

###### 중첩 질의 - ALL / ANY
예제: HADM_ID가 117105번인 환자가 갖고 있는 (모든/어떤) 체온값보다 높은 체온값을 갖고 있는 HADM_ID 추출

```python
# 모든 - ALL
q("""
select distinct hadm_id from labevents where itemid='50825' 
and value > ALL (select value from labevents where itemid='50825' and hadm_id=117105);
""")
```

```python
# 어떤 - ANY
q("""
select distinct hadm_id from labevents where itemid='50825' 
and value > ANY (select value from labevents where itemid='50825' and hadm_id=117105);
""")
```

# SQL 내장 함수
의학 데이터에서는 날짜 및 시간 함수, 수학 함수, 문자열 함수 세 가지가 유용하게 사용된다.

## 수학 함수
ABS, ROUND 등 - 슬라이드 참고

예제 1: -78, 78의 절댓값

```python
q("select abs(-78), abs(78);")
```

예제 2: 4.875를 첫째 자리까지 반올림한 값

```python
q("select round(4.875, 1)")
```

예제 3: 입원별 평균 체온 값을 구한 후 소수점 둘째자리까지 반올림한 값

```python
q("""select hadm_id, ROUND(sum(value)/count(*),2) as avg_BT
from labevents where itemid='50825' group by hadm_id;""")
```

## 문자 함수
REPLACE, LEN(length), SUBSTRING, LOWER 등 - 슬라이드 참고

예제1: 진단명에‘diabetes mellitus’라는 문구를‘DM’으로 변환하여 보이시오

```python
q("""select long_title, replace(lower(long_title), 'diabetes mellitus', 'DM') as Dx from d_icd_diagnoses
where long_title like '%diabetes mellitus%';""")
```

예제 2: d_icd_diagnoses 테이블에서 long_title들의 글자수를 출력하시오.

```python
q("select long_title, length(long_title) as Dx_len from d_icd_diagnoses;")
```

예제 3: 진단내역 테이블에서 ICD9 code의 두 자리별 처방된 수를 구하시오

```python
q("select substring(icd9_code,1,2), count(*) from d_icd_diagnoses group by substring(icd9_code,1,2);")
```
